```python
with open('../data/trec-tasks.txt') as f:
    trec = f.read()
    
trec
```

```python
import re
trec = re.sub(r'[0-9]+. ', '', trec)
trec = trec.replace("(", "")
trec = trec.replace(")", "")
trec = trec.replace("?", "")
trec = trec.replace("[", "")
trec = trec.replace("]", "")
tasks = trec.split('\n')
# tasks = [(task.split('\n\n')[0], task.split('\n\n')[1].split('\n')) for task in tasks]
tasks = list(filter(None, tasks))
tasks
```

```python
import random
new_q = random.sample(tasks, 30)
new_q
```

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
import math
```

```python
def caterr(nue):
    try:
        return 10*((1/nue)-1)
    except ZeroDivisionError:
        return 100
```

```python
similarity_mat = [[nlp(k).similarity(nlp(i))
                   for k in new_q] for i in new_q]
```

```python
similarity_mat
```

```python
distance_mat = [[caterr(k) for k in i] for i in similarity_mat]
```

```python
distance_mat
```

```python
# import csv
# with open("../data/Universal_Pool.csv", "w") as f:
#     writer = csv.writer(f)
#     writer.writerows(new_q)

# with open('../data/Distance_Matrix.csv', 'w') as f:
#     writer = csv.writer(f)
#     writer.writerows(distance_mat)
    
```

```python
from scipy.cluster.hierarchy import ward, fcluster
import numpy as np
```

```python
dist = ward(np.array(distance_mat))
clus = fcluster(dist, t=9, criterion='distance')
```

```python
clus = list(clus)
dicto = {}
for i in range(len(clus)):
    if str(clus[i]) in dicto:
        dicto[str(clus[i])].append(new_q[i])
    else:
        dicto[str(clus[i])] = [new_q[i]]
    
# print(dicto)
import json
print(json.dumps(dicto, indent=4))
```
